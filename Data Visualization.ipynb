```python
# Import important modules
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print("Setup Complete")
```

```python
# Path of the file to read
fifa_filepath = ".../fifa.csv"

# Read the file into a variavle fifa_data
fifa_data = pd.read_csv(fifa_filepath, index_col="Date", parse_dates=True)
```

```python
# Print the first five rows
fifa_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(16,6))

# line Chart showing how FIFA rankings evolved over time
sns.lineplot(data=fifa_data)
```

```python
# Load data from path
spotify_filepath = "...spotify.csv"

# Read data from file
spotify_data = pd.read_csv(spotify_filepath, index_col="Date", parse_dates=True)
```

```python
# Read first five lines of data
spotify_data.head()
```

```python
# Print last five rows of the data
spotify_data.tail()
```

```python
# Set width and height of the figure
plt.figure(figsize=(16,6))

# Add a title to the figure
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# line chart showing daily global streams of each song
sns.lineplot(data=spotify_data)
```

```python
# Get the list of columns in the data
list(spotify_data.columns)
```

```python
# Subplot of the data
# Set the width and height of the figure
plt.figure(figsize=(14,6))

# Add title
plt.title("Daily Global Streams of Popular Songs in 2017-2018")

# Line chart showing daily global streams of 'Shape of You'
sns.lineplot(data=spotify_data['Shape of You'], label="Shape of You")

# Line chart showing global streams of 'Despacito'
sns.lineplot(data=spotify_data['Despacito'], label='Despacito')

# Add label for horizontal axis and vertical axis
plt.xlabel("Date")
plt.ylabel("Number of Streams")
```

```python
# Path of the file
flight_filepath = "...flight_delays.csv"

# Read the file into a variable flight_data
flight_data = pd.read_csv(flight_filepath, index_col = "Month")
```

```python
# Read first five row of data
flight_data.head()
```

```python
# Set the width and height of the figure
plt.figure(figsize=(10,6))

# Add a title
plt.title("Average Arrival Delay for Spirit Airlines Flights, by Month")

# Barchart showing average arrival delay for Spirit Airlines flights by month
sns.barplot(x=flight_data.index, y=flight_data['NK'])

# Add a label for the vertical axis
plt.ylabel("Arrival Delay (in minutes)")
```

```python
# Set the width and height of the figure
plt.figure(figsize=(14,7))

# Add a title
plt.title("Average Arrivsl Delay for Each Airline, by Month")

# Heatmap showing average arrival delay for each airline by month
sns.heatmap(data=flight_data, annot=True)

# Add a label for the horizontal axis
plt.xlabel("Airline")
```

```python
# Path of the file to read
insurance_filepath = ".../insurance.csv"

# Read the file into a variable called insurance_data
insurance_data = pd.read_csv(insurance_filepath)
```

```python
# Read the first five rows of the data
insurance_data.head()
```

```python
# A scatterplot of BMI and Insurance charges
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'])
```

```python
# Add a regression line to the scatterplot
sns.regplot(x=insurance_data['bmi'],y=insurance_data['charges'])
```

```python
# Color coded scatterplot
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['charges'],
               hue=insurance_data['smoker'])
```

```python
# Add two regression lines corresponding to smokers and non smokers
sns.lmplot(x='bmi', y='charges', hue='smoker', data=insurance_data)
```

```python
# Categorical scatter plot
sns.swarmplot(x=insurance_data['smoker'],
             y=insurance_data['charges'])
```
